```python
import open3d as o3d
import numpy as np
import os

# Bypass WebRTC issues sometimes caused by Jupyter environments in VS Code
os.environ["WEBRTC_IP"] = "127.0.0.1"

# Define the correct path relative to the notebook's current location
point_cloud_path = "dataset/L001.ply"

# Toggles for processing and visualization
ENABLE_DOWNSAMPLING = False
ENABLE_COLORS = True  

# Load the point cloud data
print("Loading the massive point cloud... (Please wait)")
pcd = o3d.io.read_point_cloud(point_cloud_path)
processed_pcd = pcd

# Apply downsampling only if enabled
if ENABLE_DOWNSAMPLING:
    print("Downsampling is ENABLED. Reducing point cloud size...")
    processed_pcd = processed_pcd.voxel_down_sample(voxel_size=0.5)
else:
    print("Downsampling is DISABLED. Using full resolution...")

# Fix the large coordinate rendering issue by centering the point cloud
center_point = processed_pcd.get_center()
points_array = np.asarray(processed_pcd.points)
centered_points = points_array - center_point
processed_pcd.points = o3d.utility.Vector3dVector(centered_points)

# Handle color toggle
if not ENABLE_COLORS:
    print("Colors are DISABLED. Painting dark grey...")
    processed_pcd.paint_uniform_color([0.2, 0.2, 0.2])

# Render the point cloud with a custom visualizer to control point size
print("Opening the 3D visualization window...")
vis = o3d.visualization.Visualizer()
vis.create_window(window_name="High-Res Toronto-3D L001", width=1280, height=720)
vis.add_geometry(processed_pcd)

# Access render options and reduce the point size for a finer, non-blocky look
opt = vis.get_render_option()
opt.point_size = 1.0  # Decrease this value (e.g., to 0.5) if it still looks blocky

# Run the visualizer loop
vis.run()
vis.destroy_window()
```
